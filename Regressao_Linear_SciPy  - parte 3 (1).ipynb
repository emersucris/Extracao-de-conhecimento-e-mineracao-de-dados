```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importando os dados de um arquivo CSV chamado 'Regresao_Linear.csv' utilizando a biblioteca Pandas.
dados = pd.read_csv('Regresao_Linear.csv')
```

```python
#exibindo as primeiras linhas 
dados.head()
```

```python
# extraindo os valores x e y como array
X = dados['X'].values
Y = dados['Y'].values
```

```python
# criando os de dispesoes com os dados 
plt.scatter(X,Y,label='Y(X)');
plt.xlabel('X');
plt.ylabel('Y');
plt.legend();
```

```python
from scipy.stats import linregress
```

```python
# Realize a análise de regressão linear utilizando a função linregress.
"'coef_angular' e o angular da linha de ajuste."
"'coef_linear' denota o coeficiente linear da linha de ajuste."
"'std_err' corresponde ao desvio padrão do erro."
coef_angular, coef_linear,_,_,std_err = linregress(X, Y)
```

```python
print("Coeficiente angular = {:0.2f}".format(coef_angular))
```

```python
print("Coeficiente linear = {:0.2f}".format(coef_linear))
```

```python
print("Desvio padrão = {:0.2f}".format(std_err))
```

```python
# calculando a reta de ajuste 
reta = coef_angular*X+coef_linear
```

```python
# Crie um gráfico de dispersão com os dados e a reta de ajuste
plt.scatter(X,Y,label='Y(X)');
plt.plot(X,reta,label='Ajuste linear',color='red');
plt.xlabel('X');
plt.ylabel('Y');
plt.legend();
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

MAE = mean_absolute_error(Y,reta)
RMSE = np.sqrt(mean_squared_error(Y,reta))

print("MAE = {:0.2f}".format(MAE))
print("RMSE = {:0.2f}".format(RMSE))
```
